```python
import folium
import pandas as pd
import branca.colormap as cm
```

```python
df = pd.read_json('density.json')
```

```python
df.head(10)
```

```python
linear = cm.linear.Paired_06.to_step(20).scale(0.1, 80)
```

```python
m = folium.Map(location=[66, 100], zoom_start=3)

for all in range(len(df)):
    folium.Polygon(df.loc[all, 'geometries']['coordinates'], color = linear(df.loc[all, 'density']), popup = str(df.loc[all, 'density']) + 'чел./км', fill = True, weight=0.5, opacity=1).add_to(m)

folium.LayerControl().add_to(m)

m.save('russia_population.html')
m
```
